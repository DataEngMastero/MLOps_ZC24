```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
```

# Question 1 : Number of Columns in Yellow Taxi Jan 2023

```python
df = pd.read_parquet('../data/yellow_tripdata_2023-01.parquet')
df
```

```python
len(df.columns)
```

# Question 2: Calculating Duration Column's Standard Deviation

```python
df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
df.duration = df.duration.apply(lambda td: td.total_seconds() / 60)
df
```

```python
df['duration'].std() 
```

# Question 3: Dropping Outliers and Find Percentage of remaining records

```python
df['duration'].describe(percentiles=[0.95, 0.98, 0.99])
```

```python
((df.duration >= 1) & (df.duration <= 60)).mean()
```

```python
df = df[(df.duration >= 1) & (df.duration <= 60)]
df.head()
```

# Question 4: One-hot Encoding

```python
categorical = ['PULocationID', 'DOLocationID']
numerical = ['trip_distance']

df[categorical] = df[categorical].astype(str)
df[categorical].dtypes
```

```python
train_dicts = df[categorical + numerical].to_dict(orient='records')
train_dicts
```

```python
dv = DictVectorizer()
X_train = dv.fit_transform(train_dicts)
X_train
```

```python
print("Shape of X_train:", X_train.shape)
```

# Question 6: Train Model with Linear Regression and Find RSME

```python
target = 'duration'
y_train = df[target].values
y_train
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred =lr.predict(X_train)
```

```python
mean_squared_error(y_train, y_pred, squared=False)
```
